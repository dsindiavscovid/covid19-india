```python
import numpy as np
```

```python
def Lj(beta, li, xij, yj):
    w = np.exp(-beta*np.array(li))
    s = sum(w)
    w /=s
    sumwixi = np.dot(w, xij)
    return (yj - sumwixi)**2
```

```python
Lj(0.4, [6.3, 6.5, 6.7, 7, 8, 9], [128, 131, 126, 110, 180, 190], 124)
```

```python
L =[]
BETA = []
for trial in range(10):
    LjL = []
    betaL = []
    li = np.random.uniform(low=2., high=10, size=(10,))
    xij = np.random.uniform(low=110, high=140, size=(10,))
#     xij = 100*li
    yj = np.random.uniform(low=110, high=160)
    print("*************{}***********".format(trial))
    print(li)
    print(xij)
    print(yj)
    for b in range(100, 1000):
        beta = (b+0.)/100
        betaL.append(beta)
        LjL.append(Lj(beta, li, xij, yj))
    L.append(LjL)
    BETA.append(betaL)
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(10):
    print("*************{}***********".format(i))
    LjL = L[i]
    betaL = BETA[i]
    plt.clf()
    plt.plot(betaL, LjL)
    plt.show()
```
